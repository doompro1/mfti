# Reversed integer

Ğ’Ğ°Ğ¼ Ğ´Ğ°Ğ½Ğ¾ Ñ‡Ğ¸ÑĞ»Ğ¾ Ñ†ĞµĞ»Ğ¾Ğ³Ğ¾ Ñ‚Ğ¸Ğ¿Ğ° ``a``. Ğ’Ğ°Ğ¼ Ğ½ĞµĞ¾Ğ±Ñ…Ğ¾Ğ´Ğ¸Ğ¼Ğ¾ Ğ½Ğ°Ğ¿Ğ¸ÑĞ°Ñ‚ÑŒ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ, ĞºĞ¾Ñ‚Ğ¾Ñ€Ğ°Ñ Ğ¿Ñ€Ğ¸Ğ½Ğ¸Ğ¼Ğ°ĞµÑ‚ ÑÑ‚Ğ¾ Ñ‡Ğ¸ÑĞ»Ğ¾ Ğ½Ğ° Ğ²Ñ…Ğ¾Ğ´, Ğ° Ğ½Ğ° Ğ²Ñ‹Ñ…Ğ¾Ğ´ Ğ²Ñ‹Ğ´Ğ°ĞµÑ‚ Ñ†ĞµĞ»Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾, Ñ†Ğ¸Ñ„Ñ€Ñ‹ Ğ² ĞºĞ¾Ñ‚Ğ¾Ñ€Ğ¾Ğ¼ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ñ‹ Ğ² Ğ¾Ğ±Ñ€Ğ°Ñ‚Ğ½Ğ¾Ğ¼ Ğ¿Ğ¾Ñ€ÑĞ´ĞºĞµ.


**Sample Input 1:**

``

**Sample Input 2:**

``

**Sample Input 3:**

````python
def reverse_number(n: int):
    'Ğ¤ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ¿Ñ€Ğ¸Ğ½Ğ¸Ğ¼Ğ°ĞµÑ‚ Ñ‡Ğ¸ÑĞ»Ğ¾, Ğ² Ğ²Ğ¾Ğ·Ğ²Ñ€Ğ°Ñ‰Ğ°ĞµÑ‚ Ğ·ĞµÑ€ĞºĞ°Ğ»ÑŒĞ½Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾, Ğ² ĞºĞ¾Ñ‚Ğ¾Ñ€Ğ¾Ğ¼ Ñ†Ğ¸Ñ„Ñ€Ñ‹ Ğ·Ğ°Ğ¿Ğ¸ÑĞ°Ğ½Ñ‹ Ğ² Ğ¾Ğ±Ñ€Ğ°Ñ‚Ğ½Ğ¾Ğ¼ Ğ¿Ğ¾Ñ€ÑĞ´ĞºĞµ'
    try:
        n = int(str(abs(n))[-1::-1])
    except ValueError:
        print("Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ñ†ĞµĞ»Ğ¾Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ½Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ Ğ° Ğ½Ğµ Ğ²Ğ¾Ñ‚ ÑÑ‚Ğ¾ Ğ²Ğ¾Ñ‚!")
    return n


n = int(input('Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ '))
print(reverse_number(n))
````

## Ğ’Ğ°Ñ€Ğ¸Ğ°Ğ½Ñ‚ Ğ´Ğ»Ñ ÑƒĞ´Ğ¾Ğ±Ğ½Ğ¾Ğ¹ Ğ¿Ñ€Ğ¾Ğ²ĞµÑ€ĞºĞ¸

````python
import itertools
import ast
def permutations(list_:list):
    'Ğ¤ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ¿Ñ€Ğ¸Ğ½Ğ¸Ğ¼Ğ°ĞµÑ‚ Ğ¼Ğ°ÑÑĞ¸Ğ², Ğ¸ Ğ²Ğ¾Ğ·Ğ²Ñ€Ğ°Ñ‰Ğ°ĞµÑ‚ Ğ²ÑĞµ Ğ²Ğ¾Ğ·Ğ¼Ğ¾Ğ¶Ğ½Ñ‹Ğµ Ğ¿ĞµÑ€Ğ¼ÑƒÑ‚Ğ°Ñ†Ğ¸Ğ¸ Ñ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ğ°Ğ¼Ğ¸ Ğ¼Ğ°ÑÑĞ¸Ğ²Ğ°'
    try:
        res = itertools.permutations(list_, r=None)
    except ValueError:
        print("Ğ”Ğ°Ğ¹Ñ‚Ğµ Ğ½Ğ° Ğ²Ñ…Ğ¾Ğ´ Ğ¼Ğ°ÑÑĞ¸Ğ² Ğ° Ğ½Ğµ Ğ²Ğ¾Ñ‚ ÑÑ‚Ğ¾ Ğ²Ğ¾Ñ‚!")
    except TypeError:
        print("Ğ”Ğ°Ğ¹Ñ‚Ğµ Ğ½Ğ° Ğ²Ñ…Ğ¾Ğ´ Ğ¼Ğ°ÑÑĞ¸Ğ² Ğ° Ğ½Ğµ Ğ²Ğ¾Ñ‚ ÑÑ‚Ğ¾ Ğ²Ğ¾Ñ‚!")
    return res
    
list_ = ast.literal_eval(input('Ğ’ÑÑ‚Ğ°Ğ²ÑŒÑ‚Ğµ Ğ¼Ğ°ÑÑĞ¸Ğ² '))
for _ in permutations(list_):
    print(list(_))
````

## Ğ’Ğ°Ñ€Ğ¸Ğ°Ğ½Ñ‚ Ğ³Ğ´Ğµ Ğ¿Ğ¾Ğ´Ğ°Ñ‘Ñ‚ÑÑ Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ÑƒÑ Ğ² Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ¼Ğ°ÑÑĞ¸Ğ²

````python
import itertools
import ast


def permutations(list_: list):
    'Ğ¤ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ¿Ñ€Ğ¸Ğ½Ğ¸Ğ¼Ğ°ĞµÑ‚ Ğ¼Ğ°ÑÑĞ¸Ğ², Ğ¸ Ğ²Ğ¾Ğ·Ğ²Ñ€Ğ°Ñ‰Ğ°ĞµÑ‚ Ğ¼Ğ°ÑÑĞ¸Ğ² Ğ¸Ğ· Ğ²ÑĞµÑ… Ğ²Ğ¾Ğ·Ğ¼Ğ¾Ğ¶Ğ½Ñ‹Ñ… Ğ¿ĞµÑ€Ğ¼ÑƒÑ‚Ğ°Ñ†Ğ¸Ğ¹ Ñ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ğ°Ğ¼Ğ¸ Ğ¼Ğ°ÑÑĞ¸Ğ²Ğ°'
    try:
        res = itertools.permutations(list_, r=None)
    except ValueError:
        print("Ğ”Ğ°Ğ¹Ñ‚Ğµ Ğ½Ğ° Ğ²Ñ…Ğ¾Ğ´ Ğ¼Ğ°ÑÑĞ¸Ğ² Ğ° Ğ½Ğµ Ğ²Ğ¾Ñ‚ ÑÑ‚Ğ¾ Ğ²Ğ¾Ñ‚!")
    except TypeError:
        print("Ğ”Ğ°Ğ¹Ñ‚Ğµ Ğ½Ğ° Ğ²Ñ…Ğ¾Ğ´ Ğ¼Ğ°ÑÑĞ¸Ğ² Ğ° Ğ½Ğµ Ğ²Ğ¾Ñ‚ ÑÑ‚Ğ¾ Ğ²Ğ¾Ñ‚!")
    return res


nums = [1, 2, 3]
for _ in permutations(nums):
    print(list(_))
````

### ĞŸĞ¾Ñ‡Ñ‚Ğ¸ Ğ´Ğ²Ğ¾Ğ¹Ğ½Ğ¾Ğ¹ Ñ„Ğ°ĞºÑ‚Ğ¾Ñ€Ğ¸Ğ°Ğ»
Ğ”Ğ°Ğ½Ğ¾ Ğ½Ğ°Ñ‚ÑƒÑ€Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ $N \leqslant 100$.  Ğ’Ñ‹Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ğ¿Ñ€Ğ¾Ğ¸Ğ·Ğ²ĞµĞ´ĞµĞ½Ğ¸Ğµ Ğ²ÑĞµÑ… Ñ‡Ñ‘Ñ‚Ğ½Ñ‹Ñ… Ğ½Ğ°Ñ‚ÑƒÑ€Ğ°Ğ»ÑŒĞ½Ñ‹Ñ… Ñ‡Ğ¸ÑĞµĞ», Ğ½Ğµ Ğ¿Ñ€ĞµĞ²Ğ¾ÑÑ…Ğ¾Ğ´ÑÑ‰Ğ¸Ñ… $N$.

````python
import math
# Ğ”Ğ»Ñ Ğ»ÑĞ±Ğ¸Ñ‚ĞµĞ»ĞµĞ¹
#import functools

def almost_double_factorial_(n: int):
    'Ğ¤ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ¿Ñ€Ğ¸Ğ½Ğ¸Ğ¼Ğ°ĞµÑ‚ Ñ‡Ğ¸ÑĞ»Ğ¾, Ğ¿Ñ€Ğ¾Ğ¸Ğ·Ğ²ĞµĞ´ĞµĞ½Ğ¸Ğµ Ğ²ÑĞµÑ… Ñ‡Ñ‘Ñ‚Ğ½Ñ‹Ñ… Ğ½Ğ°Ñ‚ÑƒÑ€Ğ°Ğ»ÑŒĞ½Ñ‹Ñ… Ñ‡Ğ¸ÑĞµĞ», Ğ½Ğµ Ğ¿Ñ€ĞµĞ²Ğ¾ÑÑ…Ğ¾Ğ´ÑÑ‰Ğ¸Ñ…  ğ‘ .'
    try:
        if n > 1 and n < 101:
            res = math.prod(list(range(2, n, 2)))
            #res = (functools.reduce(lambda a, b: a * b, list(range(2, n, 2))))
            return res
        elif n == 1:
            return 0

        else:
            print('Ğ’Ğ°ÑˆĞµ ÑĞ¸Ğ»Ğ¾ Ğ½Ğµ Ğ²Ñ…Ğ¾Ğ´Ğ¸Ñ‚ Ğ² Ğ´Ğ¸Ğ°Ğ¿Ğ°Ğ·Ğ¾Ğ½ Ğ¾Ñ‚ 0 Ğ´Ğ¾ 100')
    except ValueError:
        print("Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ñ†ĞµĞ»Ğ¾Ñ‡Ğ¸ÑĞ»ĞµĞ½Ğ½Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ Ğ° Ğ½Ğµ Ğ²Ğ¾Ñ‚ ÑÑ‚Ğ¾ Ğ²Ğ¾Ñ‚!")


# Ğ¾Ğ±Ñ€Ğ°Ñ‰Ğ°Ñ Ğ²Ğ½Ğ¸Ğ¼Ğ°Ğ½Ğ¸Ğµ,Ñ‡ Ñ‚Ğ¾ Ğ² ÑĞ°Ğ¼Ğ¾Ğ¹ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ğ¸ ĞµÑÑ‚ÑŒ Ğ¿Ñ€Ğ¾Ğ²ĞµÑ€ĞºĞ° Ğ½Ğ° Ñ‚Ğ¸Ğ¿.
# ĞŸĞ¾Ğ»Ğµ Ğ¸Ğ½Ğ¿ÑƒÑ‚ Ñ‚ÑƒÑ‚ Ğ´Ğ»Ñ ÑƒĞ´Ğ¾Ğ±ÑÑ‚Ğ²Ğ° Ğ¸ ĞµÑĞ»Ğ¸ Ğ²Ñ‹ Ñ…Ğ¾Ñ‚Ğ¸Ñ‚Ğµ Ğ²Ğ²ĞµÑÑ‚Ğ¸ Ğ½Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾, Ñ‚Ğ¾ Ğ·Ğ°Ğ´Ğ°Ğ²Ğ°Ğ¹Ñ‚Ğµ ÑÑ‚Ğ¾ Ğ¿Ñ€ÑĞ¼Ğ¾ Ğ² ĞºĞ¾Ğ´Ğµ.
# Ğ¡Ğ°Ğ¼Ğ¾ Ğ·Ğ°Ğ´Ğ°Ğ½Ğ¸Ğµ Ñ‚Ğ°Ğº ÑÑ„Ğ¾Ñ€Ğ¼ÑƒĞ»Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¾.

n = int(input('Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ '))

N = almost_double_factorial_(n)
print(N)
````

### Ğ¥Ğ¸Ñ‚Ñ€Ğ°Ñ ÑĞ¾Ñ€Ñ‚Ğ¸Ñ€Ğ¾Ğ²ĞºĞ°

ĞŸÑƒÑÑ‚ÑŒ Ñƒ Ğ½Ğ°Ñ ĞµÑÑ‚ÑŒ ÑĞ»ĞµĞ´ÑƒÑÑ‰Ğ¸Ğ¹ ÑĞ¿Ğ¸ÑĞ¾Ğº, Ğ² ĞºĞ¾Ñ‚Ğ¾Ñ€Ğ¾Ğ¼ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ñ‹ -- tuple Ğ¸Ğ· ÑÑ‚Ñ€Ğ¾Ğº:

ĞœÑ‹ Ñ…Ğ¾Ñ‚Ğ¸Ğ¼ Ğ¾Ñ‚ÑĞ¾Ñ€Ñ‚Ğ¸Ñ€Ğ¾Ğ²Ğ°Ñ‚ÑŒ ÑÑ‚Ğ¾Ñ‚ ÑĞ¿Ğ¸ÑĞ¾Ğº Ğ¿Ğ¾ Ğ¿Ğ¾ÑĞ»ĞµĞ´Ğ½ĞµĞ¹ Ğ±ÑƒĞºĞ²Ğµ Ğ²Ñ‚Ğ¾Ñ€Ğ¾Ğ³Ğ¾ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ğ° ĞºĞ°Ğ¶Ğ´Ğ¾Ğ³Ğ¾ ``tuple``, Ñ‚.Ğµ. Ğ¿Ğ¾Ğ»ÑƒÑ‡Ğ¸Ñ‚ÑŒ Ñ‚Ğ°ĞºĞ¾Ğ¹ ÑĞ¿Ğ¸ÑĞ¾Ğº:

````python
items = [('one', 'two'), ('three', 'four'), ('five', 'six'), ('string', 'a')]
# ĞŸÑ€Ğ¾Ğ²ĞµÑ€ĞºĞ° Ñ‚Ğ¸Ğ¿Ğ° Ğ½Ğµ Ğ½ÑƒĞ¶Ğ½Ğ°. Ğ¥Ğ¾Ñ‚Ñ Ğ² Ñ‚ĞµĞ¾Ñ€Ğ¸Ğ¸ Ğ¼Ğ¾Ğ¶Ğ½Ğ¾ Ğ±Ñ‹Ğ»Ğ¾ Ğ±Ñ‹ Ğ¾Ñ‚Ğ»Ğ°Ğ²Ğ»Ğ¸Ğ²Ğ°Ñ‚ÑŒ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ñ‹,
# ĞºÑ‚Ğ¾Ñ€Ğ¾Ñ‹Ğµ Ğ½Ğµ ÑĞ²Ğ»ÑÑÑ‚ÑÑ ĞºĞ¾Ñ€Ñ‚ĞµĞ¶Ğ°Ğ¼Ğ¸. Ğ¡Ğ¼. Ğ¿Ñ€Ğ¾ÑˆĞ»Ğ¾Ğµ ÑƒĞ¿Ñ€Ğ°Ğ¶Ğ½ĞµĞ½Ğ¸Ğµ.


def sort_fun(_):
    'Ğ¤ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ´Ğ»Ñ ÑĞ¾Ñ€Ñ‚Ğ¸Ñ€Ğ¾Ğ²ĞºĞ¸ Ğ¿Ğ¾ Ğ¿Ğ¾ÑĞ»ĞµĞ´Ğ½ĞµĞ¹ Ğ±ÑƒĞºĞ²Ğµ Ğ²Ñ‚Ğ¾Ñ€Ğ¾Ğ³Ğ¾ Ğ°Ñ€Ğ³ÑƒĞ¼ĞµĞ½Ñ‚Ğ°'
    return _[1][-1]


items.sort(key=sort_fun)
print(items)
````

````python
def my_func(s: str, a: str, b: str):
    """Ğ²Ñ…Ğ¾Ğ´ Ğ¿Ğ¾Ğ´Ğ°ÑÑ‚ÑÑ Ñ‚Ñ€Ğ¸ ÑÑ‚Ñ€Ğ¾ĞºĞ¸ s, a, b, ÑĞ¾ÑÑ‚Ğ¾ÑÑ‰Ğ¸Ğµ Ğ¸Ğ· ÑÑ‚Ñ€Ğ¾Ñ‡Ğ½Ñ‹Ñ… Ğ»Ğ°Ñ‚Ğ¸Ğ½ÑĞºĞ¸Ñ… Ğ±ÑƒĞºĞ².
    Ğ—Ğ° Ğ¾Ğ´Ğ½Ñƒ Ğ¾Ğ¿ĞµÑ€Ğ°Ñ†Ğ¸Ñ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ Ğ·Ğ°Ğ¼ĞµĞ½ÑĞµÑ‚ Ğ²ÑĞµ Ğ²Ñ…Ğ¾Ğ¶Ğ´ĞµĞ½Ğ¸Ñ ÑÑ‚Ñ€Ğ¾ĞºĞ¸ a Ğ² ÑÑ‚Ñ€Ğ¾ĞºÑƒ s Ğ½Ğ° ÑÑ‚Ñ€Ğ¾ĞºÑƒ b.
    Ğ’Ğ¾Ğ·Ğ²Ñ€Ğ°Ñ‰Ğ°ĞµÑ‚ Ğ½ĞµĞ¾Ğ±Ñ…Ğ¾Ğ´Ğ¸Ğ¼Ğ¾Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾ Ğ¾Ğ¿ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹, Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ Ğ·Ğ°Ğ¼ĞµĞ½Ğ¸Ñ‚ÑŒ Ğ²ÑĞµ Ğ° Ğ½Ğ° b, ĞµÑĞ»Ğ¸ ÑÑ‚Ğ¾ Ğ½ĞµĞ²Ğ¾Ğ·Ğ¼Ğ¾Ğ¶Ğ½Ğ¾ Ğ¸Ğ»Ğ¸ Ñ‚Ñ€ĞµĞ±ÑƒĞµÑ‚ÑÑ Ğ±Ğ¾Ğ»ĞµĞµ 1000 Ğ¾Ğ¿ĞµÑ€Ğ°Ñ†Ğ¸Ğ¹, Ñ‚Ğ¾ Ğ²Ğ¾Ğ·Ñ€Ğ°Ñ‰Ğ°ĞµÑ‚ 'Impossible'."""
    n = 0
    r = None
    res = s
    while r == None:
        if n < 1000:
            s1 = s
            s = s.replace(a, b, 1)
            if s != s1:
                n += 1
            else:
                r = True
        else:
            n = 'Impossible'
            r = True
    if s == res:
        n = 'Impossible'
    del r
    del s
    del s1
    del res
    return n


s = input('Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ ÑÑ‚Ñ€Ğ¾ĞºÑƒ s ')
a = input('Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ ÑÑ‚Ñ€Ğ¾ĞºÑƒ a ')
b = input('Ğ’Ğ²ĞµĞ´Ğ¸Ñ‚Ğµ ÑÑ‚Ñ€Ğ¾ĞºÑƒ b ')
print(my_func(s, a, b))
````

````python
def cumsum_and_erase(A: list, erase=10):
    """Ğ­Ñ‚Ğ¾ Ñ„ÑƒĞ½ĞºÑ†Ğ¸Ñ, Ğ¿Ñ€Ğ¸Ğ½Ğ¸Ğ¼Ğ°ÑÑ‰ÑƒÑ Ğ¾Ğ´Ğ¸Ğ½ Ğ¾Ğ±ÑĞ·Ğ°Ñ‚ĞµĞ»ÑŒĞ½Ñ‹Ğ¹ Ğ°Ñ€Ğ³ÑƒĞ¼ĞµĞ½Ñ‚ A Ğ¸ Ğ¾Ğ´Ğ¸Ğ½ Ğ¾Ğ¿Ñ†Ğ¸Ğ¾Ğ½Ğ°Ğ»ÑŒĞ½Ñ‹Ğ¹ Ğ°Ñ€Ğ³ÑƒĞ¼ĞµĞ½Ñ‚ erase, Ğ¿Ğ¾ ÑƒĞ¼Ğ¾Ğ»Ñ‡Ğ°Ğ½Ğ¸Ñ Ñ€Ğ°Ğ²Ğ½Ñ‹Ğ¹ 1.
    Ğ²Ğ¾Ğ·Ğ²Ñ€Ğ°Ñ‰Ğ°ĞµÑ‚ Ğ¼Ğ°ÑÑĞ¸Ğ²  ğµ[0,â€¦,ğ‘âˆ’1] , Ğ³Ğ´Ğµ  ğµğ‘–=ğ´0+â€¦+ğ´ğ‘–  --- Ğ¼Ğ°ÑÑĞ¸Ğ² Ñ‡Ğ°ÑÑ‚Ğ¸Ñ‡Ğ½Ñ‹Ñ… ÑÑƒĞ¼Ğ¼ Ğ¼Ğ°ÑÑĞ¸Ğ²Ğ°  ğ´ ,
    Ğ¿Ñ€ĞµĞ´Ğ²Ğ°Ñ€Ğ¸Ñ‚ĞµĞ»ÑŒĞ½Ğ¾ ÑƒĞ´Ğ°Ğ»Ğ¸Ğ² Ğ¸Ğ· Ğ¼Ğ°ÑÑĞ¸Ğ²Ğ°  ğµ  Ğ²ÑĞµ ÑĞ»ĞµĞ¼ĞµĞ½Ñ‚Ñ‹, Ñ€Ğ°Ğ²Ğ½Ñ‹Ğµ erase."""
    i = 1
    b = []
    l = len(A)
    try:
        for _ in A:
            __ = sum(A[:i])
            b.append(__)
            i += 1
    except ValueError:
        print("Ğ”Ğ°Ğ¹Ñ‚Ğµ Ğ½Ğ° Ğ²Ñ…Ğ¾Ğ´ Ğ²ĞµÑ€Ğ½Ñ‹Ğµ Ñ‚Ğ¸Ğ¿Ñ‹ Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…!")
    except TypeError:
        print("Ğ”Ğ°Ğ¹Ñ‚Ğµ Ğ½Ğ° Ğ²Ñ…Ğ¾Ğ´ Ğ²ĞµÑ€Ğ½Ñ‹Ğµ Ñ‚Ğ¸Ğ¿Ñ‹ Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…!")
    try:
        b.remove(erase)
        return b
    except ValueError:
        return b


A = [5, 1, 4, 5, 14]
B = cumsum_and_erase(A, 29)
print(B)
````
